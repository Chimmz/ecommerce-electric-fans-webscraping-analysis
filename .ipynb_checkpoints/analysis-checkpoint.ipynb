```python
df.drop(columns=['id', 'category4'], inplace=True)
df.head()
```

```python
# Drop duplicates
df.drop_duplicates(inplace=True)
print('Current shape: ', df.shape)
df.head()
```

```python
df.isnull().sum()
```

```python
def show_pct_missing():
    pct_missing = np.round(df.isnull().sum()/df.shape[0] * 100).reset_index(name='percent_missing')
    sns.barplot(x=pct_missing['index'], y=pct_missing['percent_missing'], data=pct_missing)
    plt.show()
```

```python
show_pct_missing()
```

```python
df['price'].unique().tolist()
```

```python
def clean_price(price):
    if type(price) != str:
        return price
    
    clean = lambda p: float(p.replace('₦ ', '').replace(',', '').strip())
    price_has_range = len(price.split('-')) == 2
    
    if not price_has_range:
        return clean(price)
    prices = [clean(p) for p in price.split('-')]
    return np.mean(prices)
```

```python
df['price'] = df['price'].apply(clean_price).astype('float')
df['old_price'] = df['old_price'].apply(clean_price).astype('float')
df['price'].head()
```

```python
df['old_price']
```

```python
# Fill old price with current price (no discounts)
df['old_price'] = df['old_price'].fillna(df['price'])

df['discount'].fillna('0%', inplace=True)
df['discount'] = df['discount'].str.replace('%', '').astype('int')
df['discount'].head()
```

```python
show_pct_missing()
```

```python
cols_fill_zero = ['rating', 'total_reviews']
df[cols_fill_zero] = df[cols_fill_zero].fillna(0)
```

```python
df.total_reviews = df.total_reviews.astype('int')
```

```python
# Distribution of price
sns.distplot(df['price'], kde=True, bins=50);
```

```python
# Top 30 brands
print('Total brands:', df['brand'].nunique())

top_30_brands = df['brand'].value_counts().reset_index().head(30)

plt.figure(figsize=(16, 8))
sns.barplot(x='brand', y='count', data=top_30_brands)
plt.xticks(rotation=90);
```

```python
# Top 30 brands and the distribution of their fan prices
plt.figure(figsize=(18, 8))
sns.boxplot(x='brand', y='price', data=df);
plt.xticks(rotation=90);
```

```python
# Top 40 brands and their avg fan prices
avg_prices = np.round(df.groupby('brand')['price'].mean()).reset_index(name='avg_price')
avg_prices.sort_values('avg_price', ascending=False, inplace=True)

plt.figure(figsize=(16, 8))
sns.barplot(x='brand', y='avg_price', data=avg_prices.head(40))
plt.xticks(rotation=90);
```

```python
# Brands & their reviews/ratings
plt.figure(figsize=(16, 8))

brand_reviews_ratings = df.groupby('brand')[['total_reviews', 'rating']].mean()
brand_reviews_ratings = brand_reviews_ratings.sort_values('rating', ascending=False).reset_index().head(30)
brand_reviews_ratings
# sns.barplot(x='brand', y='total_reviews', hue='rating', data=brand_reviews_ratings);
```
